```python
# Rusul Al Busultan
import pandas as pd
with open("survey_nonvote.txt",'r') as file:
    content = file.readlines()
    line_count=0
    col = []
for line in content:
    line_count +=1
    words_list = line.split(",")
    if line_count == 1:
        ind = words_list.index("Q5")
        continue
    col.append(words_list[ind])

q5 = pd.Series(col).value_counts()

ind=["Election winner matters","Election winner does NOT matter","No Response"]

q5_pd = pd.DataFrame({' ':[q5[0], q5[1], q5[2]]},index=ind)
print(q5_pd.index)

q5_pd[' '].plot.pie(title="Survey Question Response", legend=False, \
                   autopct='%1.1f%%', 
                    startangle=0)
```
